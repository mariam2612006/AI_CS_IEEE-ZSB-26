```python
import numpy as np
```

```python
arr_2t032=np.arange(2,34,2).reshape(4,4)
mean=arr_2t032.mean()
std=arr_2t032.std()
print(mean)
print(std)
```

```python
lower =mean-0.5*std
upper=mean+0.5*std
```

```python
result=arr_2t032[(arr_2t032>=lower)&(arr_2t032<=upper)]
print(result)
```

```python
print("arr\n",arr_2t032)
print("mean:",mean)
print('std:',std)
```
